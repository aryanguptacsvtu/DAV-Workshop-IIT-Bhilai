```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
leaves = pd.read_csv("leaves_training_final.csv",sep=',')
leaves.head()
```

```python
leaves.describe()
```

```python
leaves.plot.scatter(x='width',y='length')
```

```python
sns.displot(data = leaves ,x='width',y="length",hue='edge')
```

```python
# Display basic info
print("Dataset Info:\n")
print(leaves.info())
print("\nFirst 5 rows:\n")
print(leaves.head())
```

```python
# Count of each species
plt.figure(figsize=(12, 6))
sns.countplot(data=leaves, x='name', order=leaves['name'].value_counts().index)
plt.title('Distribution of Leaf Species')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Histogram of length and width
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
sns.histplot(leaves['length'], bins=20, kde=True)
plt.title('Distribution of Leaf Length')

plt.subplot(1, 2, 2)
sns.histplot(leaves['width'], bins=20, kde=True)
plt.title('Distribution of Leaf Width')

plt.tight_layout()
plt.show()
```

```python
# Heatmap of correlation between numerical features
plt.figure(figsize=(6, 4))
sns.heatmap(leaves[['length', 'width']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

# Exercise 1 :-

### Functions

```python
#Function to calculate Cube
def cube(number):
    return number * number * number

# Using the functions
print("Cube of 4:", cube(4))
print("Cube of -2:", cube(-2))
```

```python
#Function to calculate Average
def average(nums):
  return sum(nums)/len(nums)

# Using the functions
l = list(range(1,11))
print("List : ",l)
print("Average of list : ",average(l))
```

# Exercise 2 :-

### List Comprehensions

```python
my_list = [1, 2, 3, 4, 5]
print("Original List:",my_list)

cubes = [i ** 3 for i in my_list]
odd_numbers = [i for i in my_list if i % 2 != 0]
```

```python
print("Cubes:", cubes)
print("Odd numbers:", odd_numbers)
```

## Taking Multiple Inputs and Using map()

```python
# Taking multiple integers as input
numbers = input("Enter multiple numbers separated by spaces: ").split()
numbers = list(map(int, numbers))  # Convert each string to integer
print("Numbers:", numbers)
print("Sum of numbers:", sum(numbers))
```

```python
# Taking multiple floats
scores = input("Enter multiple scores (e.g., 85.5 90.0 78.3): ").split()
scores = list(map(float, scores))  # Convert each string to float
print("Scores:", scores)
print("Average score:", sum(scores) / len(scores))
```

```python
# Combining with error handling
try:
    values = input("Enter two integers separated by a comma: ").split(',')
    x, y = map(int, values)  # Unpack two values
    print(f"Product of {x} and {y} is {x * y}")
except ValueError:
    print("Error: Please enter valid integers separated by a comma!")
```
