## ðŸ’» Using socket :-

```python
import socket
```

```python
mysock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
mysock.connect(('icio.us', 80))
```

```python
cmd = 'GET http://icio.us/HTTP/1.0\r\n\r\n'.encode()
```

```python
mysock.send(cmd)
```

```python
while True:
  data = mysock.recv(512)
  if (len(data) < 1):
    break
  print(data.decode())

mysock.close()
```

### Assignment 1 : Find no. of count of words in text file

```python
import urllib.request

# Open the URL
f = urllib.request.urlopen('http://textfiles.com/stories/antcrick.txt')

# Initialize dictionary to hold word counts
word_counts = {}

# Loop through each line
for line in f:
    words = line.decode().strip().split()

    for word in words:
        word = ''.join(char for char in word if char.isalnum())  # Remove punctuation

        if (word):  # Ensure it's not empty
            word_counts[word] = word_counts.get(word, 0) + 1

# Print the dictionary
print("\n--- Word Counts ---")
for word, count in word_counts.items():
    print(f"{word}: {count}")
```

# ðŸ’» Using BeautifulSoup :-

```python
from bs4 import BeautifulSoup
```

```python
url = input('Enter - ')
html = urllib.request.urlopen(url).read()
soup = BeautifulSoup(html, 'html.parser')

# Retrieve all of the anchor tags
tags = soup('a')
for tag in tags:
    print(tag.get('href', None))
```

### Assignment 2 : Retrieve Tags, Attributes and Text

```python
import urllib.request
from bs4 import BeautifulSoup

# Get URL from user
url = input('Enter URL: ')

# Read and parse HTML
html = urllib.request.urlopen(url).read()
soup = BeautifulSoup(html, 'html.parser')

# Retrieve all tags
print("\n--- All Tags ---")
for tag in soup.find_all(True):  # True means all tags
    print(f"Tag: <{tag.name}>")
```

```python
# Retrieve all tags with attributes
print("\n--- Tags with Attributes ---")
for tag in soup.find_all(True):
    if tag.attrs:
        print(f"<{tag.name}>: {tag.attrs}")
```

```python
# Retrieve all text
print("\n--- All Text Content ---")
print(soup.get_text())
```
